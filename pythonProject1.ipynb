```python
"""
Based on what you’ve learned until now, create a project of your choosing (impress us with your imagination). 
This project must at least satisfy the following minimum requirements:
Use at least 3 different data types. USED
Use lists or dictionaries or tuples or sets. USED
Use loops. USED
Use functions that return an output. USED
Use conditions. USED
Use a Lambda function. USED
"""
#importing date libraries to calculate the birds age and get their picture
from datetime import date
from dateutil.relativedelta import relativedelta
from IPython.display import display
from PIL import Image

#one of the lombda uses is using it inside a function
def birdAge():
    #used the lambda to deal with unknown arguments in this case each birds age where
    #lambda returns the exact age including months and days
    return lambda age:" %d years, %d months, and %d days." % (age.years, age.months, age.days)

#lambda function to open the specified image and resize it
img = lambda image: Image.open(image).resize((400,500 ), Image.ANTIALIAS)

#took the age argument from the dictionary and feed it to our function
age=birdAge() 

#used for booking sessions with the birds
def bookingFunction(pets):
    #prompting the user to enter the name of the bird they want to hangout with
    birdName=input("which bird do you want to book?")
    #the user does not need to worry about capital letters to get the bird they want
    birdName=birdName.strip(" ").replace(" ","").lower()
    #iterating through our stored birds
    for birds in pets:
        #if we found the bird the user is looking for and the bird is not booked for a different session,
        if birdName==pets[birds]["Name"].lower() and pets[birds]["AlreadyBooked"]==False:
            #confirm the users booking
            print("------------\nBird Name: ",pets[birds]["Name"], "\nBreed:",pets[birds]["Breed"],"\nAge:",pets[birds]["age"],"\nAmount Per Session:",pets[birds]["Amount Per Session"])
            #change the birds state to "booked"
            pets[birds]["AlreadyBooked"]=True
            #return the confirmation message with the booked birds picture
            display(pets[birds]["Image"])
            return "Successfully booked a session with "+pets[birds]["Name"]+". Going to main menu..."
    #returning a message if there was no match
    return "Sorry, the bird you chose is either not an option or is busy with other clients."
        
#dicitonary containing birds and their information
pets={ 
    "bird1":{"Name":"Coco","Image":img("coco.jpg"), "Breed":"Cockatiel", 
             "age":age(relativedelta(date.today(), date(2021, 1, 15))), 
             "AlreadyBooked": False, "Amount Per Session": 20.0},
    "bird2":{"Name":"Sahab","Image":img("sahab.jpg"), "Breed":"Budgie", 
             "age":age(relativedelta(date.today(), date(2020, 9, 3))), 
             "AlreadyBooked": False, "Amount Per Session": 15.0},
    "bird3":{"Name":"Mocha", "Image":img("mocha.jpg"), "Breed":"Cockatiel", 
             "age":age(relativedelta(date.today(), date(2021, 5, 23))), 
             "AlreadyBooked": False, "Amount Per Session": 20.0},
    "bird4":{"Name":"Bgdonis", "Image":img("bgdonis.jpg"), "Breed":"Cockatiel", 
             "age":age(relativedelta(date.today(), date(2023, 2, 20))), 
             "AlreadyBooked": False, "Amount Per Session": 20.0},
    "bird5":{"Name":"Beso", "Image":img("beso.jpg"), "Breed":"Green Cheek Conure", 
             "age":age(relativedelta(date.today(), date(2022, 5, 1))), 
             "AlreadyBooked": False, "Amount Per Session": 25.0},
    "bird6":{"Name":"Matcha", "Image":img("matcha.jpg"), "Breed":"Cockatoo", 
             "age":age(relativedelta(date.today(), date(2015, 7, 13))), 
             "AlreadyBooked": True, "Amount Per Session": 40.0},    
}

#welcome page
print("************Welcome to Bird Hangout Web Page************")
print("Please choose the service you want:\n1.view available birds.\n2.book a session.\n3.Exit.")

#letting the user choose one of the three options
userInput=input()
#cleaning the user input
userInput=userInput.strip(" ").replace(" ","")
#if the user did not type "3" (Exit), put them in a while loop to continue 
while userInput != "3":
    #if the user did not type the correct input
    if len(userInput)!=1:
        print("please only enter one of the specified numbers.")
    #choosing option number one (showing available birds)   
    elif len(userInput)==1 and userInput == "1":
        print("********** Showing Available Birds **********")
        #put the dictionary in a loop to show all the available birds
        for birds in pets:
            if pets[birds]["AlreadyBooked"]==False:
                print("------------\nBird Name: ",pets[birds]["Name"], "\n")
                display(pets[birds]["Image"])
                print("\nBreed:",pets[birds]["Breed"],"\nAge:",pets[birds]["age"],
                      "\nAmount Per Session:",pets[birds]["Amount Per Session"])
    #if the user typed number two (book a session) move them to the "bookingFunction" function
    elif len(userInput)==1 and userInput == "2":
        #store the resulting value (String) in the "result" variable
        result=bookingFunction(pets)
        print(result)
    print("-----------------------------------------\nPlease choose the service you want:\n1.view available sessions.\n2.book a session.\n3.Exit.")
    userInput=input()
    userInput=userInput.strip(" ").replace(" ","")
   #print(pets)
#if the user typed number three, exit the program
print("_______________________________________________________________________________\nThank you for using our service! we hope you have a fun and relaxing session with our birds!")
```
